### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Função para calcular o faturamento (sem frete)
def calcular_faturamento(df):
    return df['Preço'].sum()

# Calcular faturamento de cada loja
faturamento1 = calcular_faturamento(loja1)
faturamento2 = calcular_faturamento(loja2)
faturamento3 = calcular_faturamento(loja3)
faturamento4 = calcular_faturamento(loja4)

# Exibir os resultados sem pontuação (apenas vírgula para decimal)
print(f"Faturamento Loja 1: R$ {faturamento1:.2f}".replace('.', ','))
print(f"Faturamento Loja 2: R$ {faturamento2:.2f}".replace('.', ','))
print(f"Faturamento Loja 3: R$ {faturamento3:.2f}".replace('.', ','))
print(f"Faturamento Loja 4: R$ {faturamento4:.2f}".replace('.', ','))


```

# 2. Vendas por Categoria


```python
def vendas_por_categoria(df):
    resultado = df.groupby('Categoria do Produto')['Preço'].sum().reset_index()
    resultado.columns = ['Categoria', 'Faturamento']
    return resultado

# Aplicar para cada loja
vendas_categoria1 = vendas_por_categoria(loja1)
vendas_categoria2 = vendas_por_categoria(loja2)
vendas_categoria3 = vendas_por_categoria(loja3)
vendas_categoria4 = vendas_por_categoria(loja4)

# Exibir resultado exemplo
print("Vendas por categoria - Loja 1:")
print(vendas_categoria1)
```

```python
vendas_categoria1 = vendas_categoria1.sort_values(by='Faturamento', ascending=False)
```

# 3. Média de Avaliação das Lojas

```python
# Calcular a média das avaliações por loja
avaliacao_media1 = loja1['Avaliação da compra'].mean()
avaliacao_media2 = loja2['Avaliação da compra'].mean()
avaliacao_media3 = loja3['Avaliação da compra'].mean()
avaliacao_media4 = loja4['Avaliação da compra'].mean()

# Exibir os resultados com uma casa decimal e vírgula
print(f"Média de Avaliação Loja 1: {avaliacao_media1:.1f}".replace('.', ','))
print(f"Média de Avaliação Loja 2: {avaliacao_media2:.1f}".replace('.', ','))
print(f"Média de Avaliação Loja 3: {avaliacao_media3:.1f}".replace('.', ','))
print(f"Média de Avaliação Loja 4: {avaliacao_media4:.1f}".replace('.', ','))
```

# 4. Produtos Mais e Menos Vendidos

```python
def produtos_mais_menos_vendidos(df):
    contagem = df['Produto'].value_counts()
    mais_vendido = contagem.idxmax()
    menos_vendido = contagem.idxmin()
    return mais_vendido, contagem[mais_vendido], menos_vendido, contagem[menos_vendido]

# Loja 1
mais1, qtd_mais1, menos1, qtd_menos1 = produtos_mais_menos_vendidos(loja1)
# Loja 2
mais2, qtd_mais2, menos2, qtd_menos2 = produtos_mais_menos_vendidos(loja2)
# Loja 3
mais3, qtd_mais3, menos3, qtd_menos3 = produtos_mais_menos_vendidos(loja3)
# Loja 4
mais4, qtd_mais4, menos4, qtd_menos4 = produtos_mais_menos_vendidos(loja4)
```

```python
print(f"Loja 1 - Mais vendido: {mais1} ({qtd_mais1} vendas), Menos vendido: {menos1} ({qtd_menos1} venda)")
print(f"Loja 2 - Mais vendido: {mais2} ({qtd_mais2} vendas), Menos vendido: {menos2} ({qtd_menos2} venda)")
print(f"Loja 3 - Mais vendido: {mais3} ({qtd_mais3} vendas), Menos vendido: {menos3} ({qtd_menos3} venda)")
print(f"Loja 4 - Mais vendido: {mais4} ({qtd_mais4} vendas), Menos vendido: {menos4} ({qtd_menos4} venda)")
```

# 5. Frete Médio por Loja


```python
frete_medio1 = loja1['Frete'].mean()
frete_medio2 = loja2['Frete'].mean()
frete_medio3 = loja3['Frete'].mean()
frete_medio4 = loja4['Frete'].mean()

print(f"Frete médio Loja 1: R$ {frete_medio1:.2f}".replace('.', ','))
print(f"Frete médio Loja 2: R$ {frete_medio2:.2f}".replace('.', ','))
print(f"Frete médio Loja 3: R$ {frete_medio3:.2f}".replace('.', ','))
print(f"Frete médio Loja 4: R$ {frete_medio4:.2f}".replace('.', ','))
```
